# 문자열 실습(심화)

```python
me = "Hello World!"
print (me)
print (me[0] + " " + me[0:4]) 
# 무언가를 검색, 색인 + 리스트경우에만 대괄호, 나머지는 소괄호
# 0:4에서 4는 포함하지 않는다.
# 오른쪽 끝은 -0으로 하지 않는다. 그러면 0이 되니까

print ("Hell" in me)
# 파이썬에서 좋아하는 기능, 문구 포함여부를 체크할 수 있음

# Method를 불러오려면 변수 이후에 마침표를 찍고, tab을 누르면 리스트가 나옴
# 필요할 때마다 구글에서 검색해서 쓰면 됨.


###### ------------------------------------------
###### 처음에 특정 문자로 시작하는 것만 찾고 싶다.

# me.startswith(H)

print (me[-1])
print (me[7:-1]) # print(me[-1:7])로 하면 실행되지 않음. 거꾸로 해야 함
```

# 리스트

```python
# 변수 여러개가 있을 때, 일일이 처리하지 않고, 모아서 한꺼번에 처리할 수 있는 기능

odd = [1, 3, 5, 7, 9]
print(odd)
print(1 in odd)
print(odd[1])

odd.append(11) # append다음에 [] 대괄호가를 쓰면 작동하지 않음, ()소괄호를 써야 함
odd.append(11)
print(odd)

odd.reverse()
print(odd)
```

# 제어문(if문)

```python
age_data = [3, 10, 25, 40]

for age in age_data:
    if age < 5:
        print ("아이")
    elif age < 18:
        print ("학생")
    elif age < 28:
        print ("청년")
    else:
        print ("기타")
```

# 반복문(for문)

```python
# 리스트 안에 있는 것들을 지정한 변수에 하나하나 할당해라.

basket = ["사과", "자몽", "레몬", "포도"]

for i in basket:
    print(i)
print("\n")

for i in range(5):
    print(i)
print("\n")
    
for i in range(1,5):  # (1:5) 모양으로 가면 에러 생김.. 5번째는 빠짐
    if i == 3:
        print ("걸렸어!")
    else:   # else:를 넣지 않으면, 3도 나오게 됨
        print(i)  
```

```python
for i in range(1,10):
    if (i % 3) == 0:
        print("빵")
    else:
        print(i)
```

# 기본 분석

```python
# 성별
sns.countplot (data = train, x = "Sex", hue = "Survived")
pd.pivot_table (train, index = "Sex", values = "Survived")
```

```python
# 등급
sns.countplot (data = train, x = "Pclass", hue = "Survived")
pd.pivot_table (train, index = "Pclass", values = {"Survived", "Fare"})
# values에는 여러개를 넣을 수 있음
```

```python
# Embarked

sns.countplot (data = train, x = "Embarked", hue = "Survived")
pd.pivot_table (train, index = "Embarked", values = {"Survived", "Fare","SibSp", "Parch"})
```

# 시각화

```python
sns.lmplot (data = train, x = "Age", y = "Fare", hue = "Survived", fit_reg = False)
```
