# 데이터시각화

- matplotlib (매트플롯라이브러리) => 강하지만, 코드량이 많아야 함

- seaborn => 작은 코드량으로 구현 가능, 그래프 중 5가지 정도만 할 수 있으면 됨
- 5개로 가는데, 그걸로 안되면, seaborn의 gallery에 가서 공부하고 해보면 됨

- 그래프 내에 값을 표기시켜보고 싶다면, 그건 seaborn에서 지원하지 않음
- 레포트 용으로 하고 싶다면, table... 이런 패키지를 쓰거나, ppt 등에서 추가하는게 좋음

```python
import pandas as pd

train = pd.read_csv("data/titanic/train.csv", index_col = "PassengerId")
```

# Seaborn

```python
%matplotlib inline

import seaborn as sns
```

# Countplot

```python
sns.countplot(data = train, x = "Embarked", hue = "Pclass")
```

# Barplot

- 숫자데이터가 있으면 barplot을 쓰는 것을 추천
- y데이터가 숫자데이터이고, x데이터의 값이 서로간의 연속성이 없는 경우 (Embarked처럼)

```python
sns.barplot(data = train, x = "Embarked", y = "Age", hue = "Survived")
```

```python
sns.barplot(data = train, x ="Embarked", y ="Fare", hue ="Survived")
```

# Pointplot

- Barplot이랑 비슷하지만, y데이터가 숫자데이터이고, x데이터값이 서로 연속성이 있는
- 날짜 등

```python
sns.pointplot (data = train, x = "Embarked", y ="Fare", hue ="Survived")
```

# lmplot

- 엘엠플롯
- 회귀선은 시간이 많이 걸려서 잘 안그림 (실제로 효과도 별로 없고)
- ln [별표]로 되어 있으면 이것은 실행중이라는 의미인데, 오래감

```python
sns.lmplot (data = train, x = "Age", y = "Fare", hue ="Survived", fit_reg = False)
```

```python
low_fare = train[train["Fare"] <= 500]

print(train.shape, low_fare.shape)

sns.lmplot(data = low_fare, x = "Age", y = "Fare", hue = "Survived")
```

# distplot

- 데이터의 분포를 보여주는 것

```python
# sns.distplot (data = train, x = "Fare") 라고 하면 에러가 생김

sns.distplot(train["Fare"])

# 바 형태는 히스토그램보다는 선만 그려진 것이 초심자에게 좋음
# bin (x축의 히스토그램)... bin start와 bin...에 따라서 다르게 보이기 때문
```

```python
sns.distplot(train["Fare"], hist = False)
```

- 정규분포 or 가우시안분포라고 함
- 그 모양이 아니면, 뭔가 의미를 살펴볼 필요가 있다.

```python
low_fare2 = train[(train["Fare"] < 500) & (train["Pclass"] == 1)] 
# 등호 1개가 아니라, 2개를 써야 함

print(train.shape, low_fare2.shape)

sns.distplot(low_fare2["Fare"], hist = False)
```
