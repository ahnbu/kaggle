# WindSpeed 채우기(train)

```python
sns.countplot(data = train, x = "windspeed")
```

```python
test_wind = train.loc[train["windspeed"] == 0]
train_wind = train.loc[train["windspeed"] != 0]
train_wind.shape
test_wind.shape
test_wind.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# model = DecisionTreeClassifier(max_depth = 8, random_state = 0)
# model = DecisionTreeClassifier()
# model = DecisionTreeRegressor(random_state = 79)

model_wind = RandomForestRegressor(random_state = 79)
model_wind
```

```python
feature_wind = ['season', 'weather', 'temp',
       'atemp', 'humidity', 'year', 'month', 'day']

label_wind = "windspeed"

X_train_wind = train_wind[feature_wind]
X_test_wind = test_wind[feature_wind]
y_train_wind = train_wind[label_wind]

print(X_train_wind.shape)

model_wind.fit(X_train_wind, y_train_wind)
```

```python
predictions_wind = model_wind.predict(X_test_wind)
submission_wind = test_wind[:]
submission_wind["windspeed"] = predictions_wind

submission_wind.head(10)
submission_wind.shape
```

```python
print(train_wind.shape, submission_wind.shape)

train.shape
train_wind.head()

# 합쳐서 train 새로 업데이트
train_wind.append(submission_wind)
train = pd.concat([train_wind, submission_wind])
print(train.shape)
train.head()
train.loc[train["windspeed"] == 0] # 0이 있는지 체크
```

# WindSpeed 채우기(test)

```python
test_wind2 = test.loc[test["windspeed"] == 0]
train_wind2 = test.loc[test["windspeed"] != 0]

print(train_wind2.shape, test_wind2.shape)

model_wind2 = RandomForestRegressor(random_state = 79)
model_wind2

X_train_wind2 = train_wind2[feature_wind]
X_test_wind2 = test_wind2[feature_wind]
y_train_wind2 = train_wind2[label_wind]

print(X_train_wind2.shape)

model_wind2.fit(X_train_wind2, y_train_wind2)
```

```python
predictions_wind2 = model_wind2.predict(X_test_wind2)
submission_wind2 = test_wind2[:]

submission_wind2["windspeed"] = predictions_wind2

submission_wind2.head(10)
# submission_wind2.shape
```

# 회원, 비회원 구분 예측 (비회원)

```python
model_casual = RandomForestRegressor(random_state = 79)
model_casual
```

```python
features_name = ['season', 'holiday', 'workingday', 'weather', 'temp',
       'atemp', 'humidity', 'year', 'month', 'day', 'hour', 'dayofweek']

label_casual = "casual"

X_train_casual = train[features_name]
X_test_casual = test[features_name]
y_train_casual = train[label_casual]

print(X_train_casual.shape)

X_train_casual.head(50)

model_casual.fit(X_train_casual, y_train_casual)
```

```python
predictions_casual = model_casual.predict(X_test_casual)
submission = pd.read_csv("data/bike/sampleSubmission.csv")

submission["casual"] = predictions_casual

submission.head(10)
print(submission.shape)
```

# 회원 포함

```python
model = RandomForestRegressor(random_state = 79)
model
```

```python
label_name = "registered"

X_train = train[features_name]
X_test = test[features_name]
y_train = train[label_name]

print(X_train.shape)
X_train.head()

model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
submission["registered"] = predictions

submission["count"] = submission["casual"] + submission["registered"]

print(submission.shape)
submission.head(10)
```

```python
submission = submission[["datetime","count"]]
print(submission.shape)
submission.head(10)
```

# MAE 측정

```python
from sklearn.model_selection import cross_val_predict

y_predict = cross_val_predict(model, X_train, y_train, cv = 20)

# train 데이터 내에서 대여량 예측값이 나옴

print(y_predict.shape)
y_predict

score = np.abs(y_predict - y_train).mean()
score

print(f"Score = {score}") # 신버전 파이썬만
print(f"Score = {score:.5f}") # 소수점 5짜리까지
print("Score = {:.5f}".format(score)) # 구버전인 경우.. 추천하지는 않음

# 할 때마다 점수가 다를 수도 있는데, 
# 이는 degression의 prediction을 random을 고정시키는 게 좋다.
```
