```python
import pandas as pd
from itertools import cycle

#pd.set_option('display.unicode.east_asian_width', True)

train = pd.read_csv("data/titanic/train.csv")

from pandas import DataFrame, Series

Pclass_onehot = pd.get_dummies(train["Pclass"], prefix="Pclass")
Pclass_onehot

# Pclass로 시작하는 변수로 만들어서 자동처리

Embarked_onehot = pd.get_dummies(train["Embarked"], prefix="Embarked")
Embarked_onehot

# Embarked로 시작하는 변수로 만들어서 자동처리

train_with_onehot = pd.concat(
    [DataFrame(train),Pclass_onehot, Embarked_onehot], axis = 1)
train_with_onehot.head()
```

```python
pd.set_option('display.unicode.east_asian_width', True)

area = '서울 부산 대구 인천 광주 대전 울산 세종 경기 강원 충북 충남 전북 전남 경북 경남 제주'.split()
college = '인문대학 사회과학대학 자연과학대학 간호대학 경영대학 공과대학 미술대학'.split()
gender = '남성 여성'.split()

# 100개의 가짜 데이터 생성, itertools.cycle 함수로 각 요소를 순환시킵니다.

fake_data = zip(range(100), cycle(area), cycle(college), cycle(gender))

# zip은 파이썬 내장함수, 같은 개수의 여러개의 자료형을 서로 묶어주는 역할

hundred_students = DataFrame([data for num, *data in fake_data],
                              columns='지역 단과대 성별'.split())
hundred_students.head(10)
```

```python
college_one_hot_encoded= pd.get_dummies(hundred_students["단과대"])

college_one_hot_encoded

# hundred_students["단과대"]는 hudred_students.단과대
# 로 표현할 수 있음
```

```python
college_with_onehot = pd.concat(
    [DataFrame(hundred_students["단과대"]), college_one_hot_encoded], axis = 1)
# hundred_students 뒤의 s를 주의
# axis =1 은 열로 작업할 때 표기, axis = 0은 행을 추가하여 작업할 때.
# pd.concat([df1, df2])는 df1과 df2를 합칠 때 사용

college_with_onehot.head(10)
```
