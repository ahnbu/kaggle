# 제어문(if문)

```python
age_data = [3, 10, 25, 40]

for age in age_data:
    if age < 5:
        print ("아이")
    elif age < 18:
        print ("학생")
    elif age < 28:
        print ("청년")
    else:
        print ("기타")
```

# 반복문(for문)

```python
# 리스트 안에 있는 것들을 지정한 변수에 하나하나 할당해라.

basket = ["사과", "자몽", "레몬", "포도"]

for i in basket:
    print(i)
print("\n")

for i in range(5):
    print(i)
print("\n")
    
for i in range(1,5):  # (1:5) 모양으로 가면 에러 생김.. 5번째는 빠짐
    if i == 3:
        print ("걸렸어!")
    else:   # else:를 넣지 않으면, 3도 나오게 됨
        print(i)  
```

```python
for i in range(1,10):
    if (i % 3) == 0:
        print("빵")
    else:
        print(i)
```

```python
train[train["Sex"] == "male"].head()
train["Embarked"] == "S" 
# 이렇게 true, false 했을 때 맨 아래 row 개수가 embarked 개수와 동일하면 적용 가능

#########################
# Q. 이게 동일하지 않은 경우도 있는지?
 
# Name: Embarked, Length: 891, dtype: bool

train[train["Embarked"] == "S" ] # 문자
train[train["Pclass"] == 1]  # 숫자


# NaN (Not a Number), null 찾기

train[train["Age"].isnull()].head() # 비어있는 것만 
#train[train["Age"].notnull()].head() # 안비어있는 것만
```

```python
train[(train["SibSp"] == 0) & (train["Parch"] == 0)].head()
# 이렇게 하면 너무 길어지고 지저분해져서, 변수로 지정한 후 불러오는 방법 추천

# and (and라고 치면 안됨. &라고 해야 함) 
without_sibsp = (train["SibSp"] == 0)
without_parch = (train["Parch"] == 0)
train[without_sibsp & without_parch].head()

# or (or라고 치면 안됨)
train_2 = train[without_sibsp | without_parch]
print(train_2.shape) # 개수를 위에서 보고 싶을 때.
train[without_sibsp | without_parch].head()
```

# 컬럼 추가하기 & 수정하기

```python
train["Category"] = "Titanic"
train.head()

#---------------------

train["id"] = range(891)  # 개수를 맞춰줘야 함. 만약 890, 892로 넣으면 에러 발생
train["id"] = 1
train.head()

#---------------------

train["Familysize"] = train["SibSp"] + train["Parch"] + 1
train.head() # 전체 테이블
train[["SibSp","Parch","Familysize"]].head() # 3개만 보고 싶을 때
train[["Familysize","SibSp","Parch"]].head() # 넣는 순서에 따라서 보여줌

#---------------------

train["France"] = (train["Embarked"] == "C")
train["UK"] = (train["Embarked"] != "C")
train[["Embarked","France", "UK"]].head()

#---------------------
```
