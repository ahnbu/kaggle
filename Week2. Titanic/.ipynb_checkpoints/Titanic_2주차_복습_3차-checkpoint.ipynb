```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import sklearn
from sklearn.tree import DecisionTreeClassifier # sklearn이 아니라, sklearn.tree임
```

```python
train = pd.read_csv("data/titanic/train.csv", index_col = "PassengerId")
test = pd.read_csv("data/titanic/test.csv", index_col = "PassengerId")

train.describe()
test.describe()

print(train.shape, test.shape)

assert train.shape == (891, 11)
assert test.shape == (418, 10)
```

데이터 전처리(NaN)

```python
#### NaN 처리할 변수 추출

print(train.columns)

print(train.describe())
# Age

print(train[["Name", "Sex", "Ticket", "Cabin", "Embarked"]].describe())
# Cabin, Embarked

print(test.describe())
# Age, Fare

print(test[["Name", "Sex", "Ticket", "Cabin", "Embarked"]].describe())
# Cabin
```

데이터 전처리(NaN) - 정성

```python
#### Age

train["Age_encode"] = train["Age"]
train.loc[train["Age"].isnull(), "Age_encode"] = train["Age"].mean()
train.loc[train["Age"].isnull()]

test["Age_encode"] = test["Age"]
test.loc[test["Age"].isnull(), "Age_encode"] = test["Age"].mean()
test.loc[test["Age"].isnull()]

```
